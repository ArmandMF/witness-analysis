```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Base/WIP_TOBE_B1000.xlsx', sheet_name='Sheet1')
data_c1 = pd.read_excel('SC1/WIP_TOBE_B1000.xlsx', sheet_name='Sheet1')
data_c2 = pd.read_excel('SC2/WIP_TOBE_B1000.xlsx', sheet_name='Sheet1')
data_c3 = pd.read_excel('SC3/WIP_TOBE_B1000.xlsx', sheet_name='Sheet1')

data_c1_release = pd.read_excel('SC1/WIP_TOBE_release.xlsx', sheet_name='Sheet1')
```

```python
data = data.iloc[::2]
data_c1 = data_c1.iloc[::2]
data_c2 = data_c2.iloc[::2]
data_c3 = data_c3.iloc[::2]
data_c1_release = data_c1_release.iloc[::2]
```

```python
print(data.head(10))
```

```python
ax = plt.gca()

data.plot(kind='line',x='Time',y='WIP',ax=ax, figsize=(15, 9))
data.plot(kind='line',x='Time',y='WIP_A', color='red', ax=ax)
data.plot(kind='line',x='Time',y='WIP_B', color='green', ax=ax)
data.plot(kind='line',x='Time',y='WIP_C', color='cyan', ax=ax)
data.plot(kind='line',x='Time',y='WIP_D', color='magenta', ax=ax)
data.plot(kind='line',x='Time',y='WIP_E', color='yellow', ax=ax)
data.plot(kind='line',x='Time',y='WIP_F', color='black', ax=ax)
data.plot(kind='line',x='Time',y='WIP_G', color='orange', ax=ax)
data.plot(kind='line',x='Time',y='WIP_H', color='purple', ax=ax)
plt.title('WIP levels (Buffers = 1000)')
plt.xlabel('Time')
plt.ylabel('Units')
```

```python
ax = plt.gca()

#data.plot(kind='line',x='Time',y='WIP',ax=ax, )
data_c1.plot(kind='line',x='Time',y='WIP', ax=ax, figsize=(15, 9), label='WIP Scenario 1')
data_c2.plot(kind='line',x='Time',y='WIP',  ax=ax, label='WIP Scenario 2')
data_c3.plot(kind='line',x='Time',y='WIP',  ax=ax, label='WIP Scenario 3')
data_c1_release.plot(kind='line',x='Time',y='WIP',  ax=ax, label='WIP Scenario ')
```
