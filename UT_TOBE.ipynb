```python
areas_c1 = {'Areas':['A1', 'A2', 'A3'], 'UT':[data_c1_sliced.mean().iloc[[0,1,3,7,9]].mean(), data_c1_sliced.mean().iloc[[2,5,8]].mean(), data_c1_sliced.mean().iloc[[4,6,10,11]].mean()]}
areas_df_c1 = pd.DataFrame(areas_c1)
areas_c1 = pd.Series(areas_df_c1['UT'], index=areas_df_c1.index)
```

```python
import seaborn as sns
sns.set_context('poster')

fig, ax = plt.subplots(figsize=(25, 8))

x = np.arange(3)

bar_width = 0.2

b0 = ax.bar(x, areas_base, width=bar_width,color="#f9ed69",label ='Base')
b1 = ax.bar(x + bar_width, areas_c1,width=bar_width,color="#f08a5d", label='Scenario 1')
b2 = ax.bar(x + 2*bar_width, areas_c2, width=bar_width,color="#b83b5e",label='Scenario 2')
b3 = ax.bar(x + 3*bar_width, areas_c3, width=bar_width,color="#6a2c70",label='Scenario 3')



# Fix the x-axes.
ax.set_xticks(x + 1.5*bar_width)
ax.set_xticklabels(('Cell 1','Cell 2','Cell 3'))

# Axis styling.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

# Add axis and chart labels.
#ax.set_xlabel('Functional areas', labelpad=15)
ax.set_ylabel('% Utilisation', labelpad=15)
ax.set_title('Utilisation of cells', pad=15)


ax.legend()

plt.legend(bbox_to_anchor=(0.96,0.6,0.15,0.2), loc="lower right", mode="expand", borderaxespad=0, ncol=1)


fig.tight_layout()
plt.savefig('UT_AREAS_TOBE.png', transparent=True, dpi=200)
```
